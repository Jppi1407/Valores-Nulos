**Gastos y costos 20-22_limpio (Hoja 2020)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_2020 = pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name="2020")
df_2020.head()
```

```python
#Eliminar filas innecesarias
data1= pd.read_excel('Gastos y costos 20-23.xlsx', skiprows=5, sheet_name='2020')
data1
```

```python
#Identificar valores nulos por columna
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
data2= data1.drop(['Columna1'], axis=1)
data2
```

```python
#Identificar valores nulos por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Crear una copia del DataFrame original
data3=data2.copy()
```

```python
#Sustituir valores nulos con promedio o media -IMPORTE-
data3["IMPORTE"]=data2["IMPORTE"].fillna(round(data2["IMPORTE"].mean(),1))
```

```python
#Identificar valores nulos por columna
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por un string en  concreto
data3["GASTO"]=data2["GASTO"].fillna("No aplica")
```

```python
#Identificar valores nulos por columna
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por un valor numérico en  concreto -TC-
data3["TC"]= data2["TC"].fillna(0)
data3
```

```python
#Identificar valores nulos por columna
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill") -FOLIO-
#Filtro por columnas
data3["FOLIO"] =data3["FOLIO"].fillna(method="ffill")

#Identificar valores nulos por columna
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por un valor numérico en  concreto -IVA-
data3["IVA"]= data3["IVA"].fillna(0)

#Identificar valores nulos por columna
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill") -TIPO-
#Filtro por columnas
data3["TIPO"] =data3["TIPO"].fillna(method="ffill")

#Identificar valores nulos por columna
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill") -POLIZA-
#Filtro por columnas
data3["POLIZA"] =data3["POLIZA"].fillna(method="ffill")

#Identificar valores nulos por columna
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
data2.to_csv("Gastos y costos 20_sin_nulos.csv")
```

**Gastos y costos 20-22_limpio (Hoja 2022)**

```python
import pandas as pd
df_2022= pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name= "2022",  skiprows=3)
print(df_2022.shape)
df_2022.head(10)
```

```python
valores_nulos=df_2022.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df_2022["Folio"] = df_2022["Folio"].fillna(method="ffill")


#Identificar valores nulos por columna
valores_nulos=df_2022.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de acuerdo a condicional de columna Descripción y MP (Pago = PAGO)
MP_NULL = df_2022['MP'].isna()
df_2022.loc[MP_NULL, 'MP'] = 'PAGO'
#Identificar valores nulos por columna
valores_nulos=df_2022.isnull().sum()
valores_nulos
```

```python
# Rellenar valores nulos de TC con 0
df_2022["TC"] = df_2022["TC"].fillna(0)

# Verificar nuevamente valores nulos
valores_nulos = df_2022.isnull().sum()
valores_nulos
```

```python
# Rellenar valores nulos de Poliza con "SIN POLIZA"
df_2022["Poliza"] = df_2022["Poliza"].fillna("SIN POLIZA")

# Verificar nuevamente valores nulos
valores_nulos = df_2022.isnull().sum()
valores_nulos
```

```python
# Eliminar la columna "Otros"
df_2022 = df_2022.drop(columns=["Otros"])

# Verificar nuevamente valores nulos
valores_nulos = df_2022.isnull().sum()
valores_nulos
```

**Gastos y costos 20-23_limpio (Hoja 2023)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carga de archivo
df_2023 = pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name="2023", skiprows=3)
print(df_2023.columns)
print(df_2023.shape)
df_2023.head(5)
```

```python
#Identificar valores nulos por columna
valores_nulos=df_2023.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
df_2023_2 = df_2023
df_2023_2.head(5)
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df_2023_2["Folio"] = df_2023_2["Folio"].fillna(method="ffill")
df_2023_2["Proveedor"] = df_2023_2["Proveedor"].fillna(method="ffill")
df_2023_2["TIPO GASTO"] = df_2023_2["TIPO GASTO"].fillna(method="ffill")

#Identificar valores nulos por columna
valores_nulos=df_2023.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de acuerdo a condicional de columna Descripción y MP (Pago = PAGO)
MP_NULL = df_2023_2['MP'].isna()
df_2023_2.loc[MP_NULL, 'MP'] = 'PAGO'

#Sustitución de valores nulos de acuerdo a condicional de columna Descripción y FP (Pago = 00)
df_2023_2["FP"]= df_2023_2["FP"].fillna(00)
df_2023_2

display(df_2023_2.isnull().sum())
```

```python
#Eliminar columna completa 'Otros' sin valores
df_2023_2 = df_2023_2.drop(["Otros"], axis = 1)
df_2023_2
```

```python
display(df_2023_2.isnull().sum())
```

```python
#Rellenar los valores nulos con nombre 'NONE' en la variable 'Poliza'
df_2023_2['Poliza'] = df_2023_2['Poliza'].fillna('NONE')
display(df_2023_2.isnull().sum())
```

```python
#Pasar dataframe limpia a CSV 'Gastos y costos 20-23_limpio.CSV'
df_2023_2.to_csv("Gastos y costos 20-23_limpio.csv", index = False)
from google.colab import files
files.download("Gastos y costos 20-23_limpio.csv")
```
