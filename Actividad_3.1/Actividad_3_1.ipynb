```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('Ventas_totales.csv')
df.head(5)
```

```python
#Identificar valores nulos por columna
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
df2 = df
df2
```

```python
#Sustituir valores nulos con mediana
df2["salon_ventas"]=df2["salon_ventas"].fillna(round(df["salon_ventas"].median(),1))
```

```python
# Calcular valores nulos por columna
valores_nulos = df.isnull().sum()
# Convertir a DataFrame para mayor legibilidad
valores_nulos = valores_nulos.reset_index()
valores_nulos.columns = ['Columna', 'Valores_Nulos']

print(valores_nulos)
```

```python
#Sustituir valores nulos por un valor numÃ©rico en  concreto
df2["tarjetas_debito"]= df["tarjetas_debito"].fillna(10)
pd.options.display.max_rows = 50
df2
```

```python
display(df2.isnull().sum())
```

```python
#Sustituir valores nulos con promedio o media
df2["tarjetas_credito"]=df2["tarjetas_credito"].fillna(round(df["tarjetas_credito"].mean(),1))
display(df2.isnull().sum())
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df2["otros_medios"] =df2["otros_medios"].fillna(method="ffill")

display(df2.isnull().sum())
```

```python
#Sustituir valores nulos con promedio o media
df2["subtotal_ventas_alimentos_bebidas"]=df2["subtotal_ventas_alimentos_bebidas"].fillna(round(df["subtotal_ventas_alimentos_bebidas"].mean(),1))

display(df2.isnull().sum())
```

```python
#Sustituir valores nulos por valores no nulos hacia atrÃ¡s backward fill" ("bfill")
#Filtro por columnas
df2["bebidas"] =df2["bebidas"].fillna(method="bfill")
df2["panaderia"] =df2["panaderia"].fillna(method="bfill")
df2["lacteos"] =df2["lacteos"].fillna(method="bfill")
df2["carnes"] =df2["carnes"].fillna(method="bfill")
df2["verduleria_fruteria"] =df2["verduleria_fruteria"].fillna(method="bfill")
df2["alimentos_preparados_rotiseria"] =df2["alimentos_preparados_rotiseria"].fillna(method="bfill")

display(df2.isnull().sum())
```

```python
#Sustituir valores nulos con promedio o media
df2["almacen"]=df2["subtotal_ventas_alimentos_bebidas"].fillna(round(df["almacen"].mean(),1))
df2["indumentaria_calzado_textiles_hogar"]=df2["indumentaria_calzado_textiles_hogar"].fillna(round(df["indumentaria_calzado_textiles_hogar"].mean(),1))
df2["electronicos_articulos_hogar"]=df2["electronicos_articulos_hogar"].fillna(round(df["almacen"].mean(),1))
df2["otros"]=df2["otros"].fillna(round(df["otros"].mean(),1))

display(df2.isnull().sum())
```

```python
df2.to_csv("Ventas_totales_limpio.csv", index = False)
```
